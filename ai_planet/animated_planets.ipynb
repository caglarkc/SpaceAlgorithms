```python
import numpy as np
import matplotlib.pyplot as plt

# Kepler'in denklemi için fonksiyon
def kepler_orbit(mean_anomaly_deg, e):
    """
    Kepler denklemini çözerek gezegenin konumunu (Eccentric Anomaly) hesaplar.
    :param mean_anomaly_deg: Ortalama anomali (derece cinsinden)
    :param e: Yörünge eliptikliği (eccentricity)
    :return: Eccentric anomaly (E) (radyan cinsinden)
    """
    M = np.radians(mean_anomaly_deg)  # Mean anomaly (dereceden radyana)
    E = M  # Başlangıç tahmini olarak E = M
    
    # Newton's method for solving Kepler's equation
    for _ in range(10):  # Daha hassas bir çözüm için iterasyon
        E = E - (E - e * np.sin(E) - M) / (1 - e * np.cos(E))

    return E

# Örnek veriler: 
# Bir gezegenin ortalama anomali ve yörünge eliptikliği (eccentricity)
mean_anomaly = 126.06756  # Derece cinsinden
eccentricity = 0.2304384  # Eccentricity

# Kepler denklemini çözerek gezegenin Eccentric Anomaly'sini hesaplayalım
E = kepler_orbit(mean_anomaly, eccentricity)

# Gezegenin yörüngesindeki konumu hesaplamak için
# Gezegenin yörüngesinin yarıçapı (r), Eccentric anomaly (E) ile hesaplanabilir
semi_major_axis = 2.7703442  # Yarıçap (AU cinsinden)

# Yörünge eliptikliğine göre gezegenin konumu (r)
r = semi_major_axis * (1 - eccentricity * np.cos(E))

# Gezegenin X ve Y koordinatlarını hesaplayalım (yörünge düzleminde)
x = r * np.cos(E)  # Gezegenin X koordinatı
y = r * np.sin(E)  # Gezegenin Y koordinatı

# Gezegenin Yörüngesinin Konumlarını görselleştirelim
plt.figure(figsize=(6, 6))
plt.plot(x, y, 'bo', label=f"Gezegen: {mean_anomaly}° Anomali")
plt.xlabel('X (AU)')
plt.ylabel('Y (AU)')
plt.title(f'Gezegenin Yörüngedeki Konumu (Eccentricity: {eccentricity})')
plt.grid(True)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()

# Hesaplanan değerleri yazdıralım
print(f"Ortalama Anomali: {mean_anomaly}°")
print(f"Eccentric Anomaly (E): {np.degrees(E)}°")
print(f"Yörüngedeki Konum (r): {r} AU")
print(f"X Koordinatı: {x} AU")
print(f"Y Koordinatı: {y} AU")
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from skyfield.api import load
import numpy as np


# Ephemeris dosyasını yükle
planets = load('de421.bsp')
ts = load.timescale()
t = ts.now()


# Güneş Sistemi'ndeki gezegenlerin isimlerini, Skyfield nesnelerini ve renklerini tanımla
planet_names = ['Sun', 'Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
planet_objects = [planets['sun'], planets['mercury'], planets['venus'], planets['earth'], planets['mars'],
                  planets['jupiter barycenter'], planets['saturn barycenter'],
                  planets['uranus barycenter'], planets['neptune barycenter']]
planet_colors = ['yellow', 'gray', 'orange', 'blue', 'red', 'brown', 'gold', 'cyan', 'purple']


# 3D koordinatlar için listeler oluştur
x_vals = []
y_vals = []
z_vals = []


# Dünya referans noktamız olacak
earth = planets['earth']


# Gezegenlerin pozisyonlarını hesapla ve 3D koordinatlara çevir
for planet in planet_objects:
    position = earth.at(t).observe(planet).apparent()
    ra, dec, distance = position.radec()
   
    # RA, DEC ve uzaklığı 3D koordinatlara çevir
    x = distance.au * np.cos(np.radians(dec.degrees)) * np.cos(np.radians(ra.hours * 15))
    y = distance.au * np.cos(np.radians(dec.degrees)) * np.sin(np.radians(ra.hours * 15))
    z = distance.au * np.sin(np.radians(dec.degrees))
   
    x_vals.append(x)
    y_vals.append(y)
    z_vals.append(z)


# Yakın yıldız sistemlerinin manuel olarak girilen pozisyonları
star_names = ['Alpha Centauri', 'Barnard\'s Star', 'Luyten\'s Star', 'Proxima Centauri']
star_ra_degrees = [219.9, 269.5, 133.0, 217.4]  # RA değerleri (derece cinsinden)
star_dec_degrees = [-60.83, 4.67, -17.99, -62.68]  # DEC değerleri (derece cinsinden)
star_distances_pc = [1.34, 5.96, 12.37, 1.29]  # Uzaklık (parsec cinsinden)


# Yıldız sistemlerinin 3D koordinatlarını hesapla
for ra, dec, dist in zip(star_ra_degrees, star_dec_degrees, star_distances_pc):
    # RA, DEC ve uzaklığı 3D koordinatlara çevir (parsec biriminden)
    x = dist * 3.26 * np.cos(np.radians(dec)) * np.cos(np.radians(ra))
    y = dist * 3.26 * np.cos(np.radians(dec)) * np.sin(np.radians(ra))
    z = dist * 3.26 * np.sin(np.radians(dec))
   
    x_vals.append(x)
    y_vals.append(y)
    z_vals.append(z)


# 3D Grafik Oluştur
fig = plt.figure(figsize=(14, 12))
ax = fig.add_subplot(111, projection='3d')


# Gezegenleri çiz ve isim etiketlerini ekle
for i, name in enumerate(planet_names):
    ax.scatter(x_vals[i], y_vals[i], z_vals[i], color=planet_colors[i], s=100, label=name)
    ax.text(x_vals[i], y_vals[i], z_vals[i], name, color=planet_colors[i])


# Yıldızları çiz ve isim etiketlerini ekle
for i, name in enumerate(star_names, start=len(planet_names)):
    ax.scatter(x_vals[i], y_vals[i], z_vals[i], color='orange', s=100, marker='*')
    ax.text(x_vals[i], y_vals[i], z_vals[i], name, color='orange')


# Eksen ayarları
ax.set_xlabel("X (AU)")
ax.set_ylabel("Y (AU)")
ax.set_zlabel("Z (AU)")
plt.title("3D View of Solar System and Nearby Star Systems")


# Eksen limitlerini ayarla (daha yakın bir görünüm için -10 ile 10 AU arasında sınırlıyoruz)
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_zlim(-10, 10)


# Kamera açısını ayarla
ax.view_init(elev=30, azim=60)


# Grafiği göster
plt.legend(loc='upper left')
plt.show()
```

```python
from skyfield.api import load

# Ephemeris dosyasını yükle
planets = load('de421.bsp')
ts = load.timescale()
t = ts.now()

# Güneş Sistemi'ndeki gezegenlerin isimlerini ve Skyfield nesnelerini tanımla
planet_names = ['Sun', 'Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
planet_objects = [planets['sun'], planets['mercury'], planets['venus'], planets['earth'], planets['mars'],
                  planets['jupiter barycenter'], planets['saturn barycenter'], 
                  planets['uranus barycenter'], planets['neptune barycenter']]

# Dünya referans noktamız olacak
earth = planets['earth']

# Veri türünü ve anahtarları döndüren fonksiyon
def get_data_structure():
    data_info = {}
    for name, planet in zip(planet_names, planet_objects):
        position = earth.at(t).observe(planet)  # Gezegenin konum verisini al
        ra, dec, distance = position.radec()    # Sağ açıklık, dik açıklık ve mesafe
        
        # Her veri türü için anahtarlar ve veri tiplerini bir sözlükte sakla
        data_info[name] = {
            "Right Ascension (RA)": (ra, type(ra)),
            "Declination (DEC)": (dec, type(dec)),
            "Distance": (distance, type(distance))
        }
    
    # Veriyi ekrana yazdır
    for planet, data in data_info.items():
        print(f"\n{planet} Data:")
        for key, (value, value_type) in data.items():
            print(f" - {key}: Value = {value}, Type = {value_type}")

# Fonksiyonu çağır
get_data_structure()
```
